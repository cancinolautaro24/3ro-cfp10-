```
class Celular:
  def __init__(self, nombre, marca, modelo):
    self.nombre=nombre
    self.marca=marca
    self.modelo=modelo
    self.bateria=100
    self.estaPrendido=False

def apagar(self):
  if(self.estaPrendido==False):
    print("esta apagado")
  else:
    self.estaPrendido=False
    print("esta apagado")

def usar_celular(self,cantidad,minutos):
  if(self.estaPrendido==True):
    if(self.bateria>cantidad_minutos):
      self.bateria-=cantidad_minutos
    else:
      self.bateria=0
      self.apagar()



```
